### Squad Standard -  Team stats

```python
team = pd.read_csv(r"/Users/goutham/Downloads/Standard.csv", header = 1)

# Set the max_columns option to None to display all columns
pd.set_option('display.max_columns', None)

# Define a dictionary of old and new column names
rename_team = {
    "Gls.1": "Gls/90",
    "Ast.1": "Ast/90",
    "G+A.1": "G+A/90",
    "G-PK.1": "G-PK.1/90",
    "G+A-PK": "G+A-PK/90",
    "xG.1": "xG/90",
    "xAG.1": "xAG/90",
    "xG+xAG": "xG+xAG/90",
    "npxG.1": "npxG./90",
    "npxG+xAG.1": "npxG+xAG/90"
    
}

# Rename the columns using the dictionary
team = team.rename(columns=rename_team)

# Display the modified DataFrame
team
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
Team = team["Squad"]
ProgressivePasses = team["PrgP"]
ProgressiveCarries = team["PrgC"]

# Define the size of the bubbles
sizes = (ProgressivePasses * ProgressiveCarries)/200

# Define the threshold values for each cluster
low_thresh = np.percentile(sizes, 30)
high_thresh = np.percentile(sizes, 75)

# Define the colors for each cluster
colors = np.where(sizes > high_thresh,'#2E8B57' , np.where(sizes < low_thresh,'#DC143C' ,'#4169E1' ))

# Create the bubble chart
fig, ax = plt.subplots(figsize=(12,9))
ax.scatter(ProgressivePasses, ProgressiveCarries, s=sizes, alpha=0.5, c=colors)

# Add labels and titles
ax.set_xlabel("Progressive Passes Made", fontsize=12)
ax.set_ylabel("Progressive Carries Made", fontsize=12)
ax.set_title("Progressive Passes and Carries Made by Premier League Teams (2022-23 season)", fontsize=16, pad=20)

# Change the font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']

# Add team labels to the bubbles with adjusted position
for i, txt in enumerate(Team):
    ax.annotate(txt, (ProgressivePasses[i], ProgressiveCarries[i]), textcoords='offset points', xytext=(4,5), fontsize=10)


# Add gridlines
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

# Set the background color
ax.set_facecolor('#F5F5F5')

# Adjust the chart layout and padding
fig.tight_layout(pad=3)


# Add your name and data source
ax.text(1, -0.07, 'Created by @gouthxm', transform=ax.transAxes, ha='right', fontsize=12.5)
ax.text(1, -0.1, 'Source: FBREF', transform=ax.transAxes, ha='right', fontsize=12.5)

# Save the chart in high resolution
fig.savefig('chart1.png', dpi=600)


```

### Calculating Percentiles

```python
import pandas as pd
from scipy.stats import rankdata

# Load the data into a pandas DataFrame


# Loop through each column in the DataFrame
for col in radar.columns:
    # Rank the data by percentile using scipy.stats.rankdata()
    ranks = rankdata(radar[col]) / len(radar[col])
    # Add the percentile ranks as a new column to the DataFrame
    radar[f'{col}_percentile'] = ranks

# Print the DataFrame with the percentile ranks
radar
```

```python
# Save the DataFrame as a CSV file
radar.to_csv('premier_league_stats_percentiles.csv', index=False)
```

#### Radar Chart

```python
import numpy as np
import matplotlib.pyplot as plt

params = ['Cmp_S_percentile', 'Cmp_M_percentile', 'Cmp_L_percentile', 'PPA_percentile', '1/3_percentile', 'KP_percentile', 'CrsPA_percentile', 'Cmp_percentile']
values = [0.75, 0.60, 0.40, 0.75, 0.70, 0.850, 0.225, 0.65]

# convert the values to percentages
values = [value*100 for value in values]

# specify the desired names for each axis
labels = ['Short Pass', 'Medium Pass', 'Long Pass', 'Pass into Penalty area', 'Final third', 'Key Passes', 'Crosses into Penalty area', 'Pass Completion']

# number of variables
num_vars = len(params)

# create an array of angles for each axis
angles = np.linspace(0, 2*np.pi, num_vars, endpoint=False)

# close the plot by repeating the first axis
values += values[:1]
angles = np.concatenate((angles,[angles[0]]))

# Define the color scheme
colors = plt.cm.viridis(np.linspace(0, 1, num_vars))

# initialize the plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

# plot the radar chart and add labels
for i in range(num_vars):
    angle = angles[i]
    value = values[i]
    label = "{:.0f}".format(value) # format the label string as percentage
    ax.text(angle, value+5, label, ha="center", va="center",  fontsize=12) # add label to the plot
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)


# set the labels for each axis using the labels list
ax.set_thetagrids(angles[:-1] * 180/np.pi, labels, fontsize = 14)

# set the range of values for each axis
ax.set_ylim(0, max(values)+10) # adjust the y-axis limit to fit the data

# Add a gray background
fig.patch.set_facecolor('#FFFACD')

# add a title
plt.title('Manchester United 22/23 passing Percentile Rank', fontsize = 15, fontweight = 'bold')

# add a grid
ax.grid(True)

# Adjust the chart layout and padding
fig.tight_layout(pad=3)


# Add your name and data source
ax.text(1, -0.07, 'Created by @gouthxm', transform=ax.transAxes, ha='right', fontsize=12.5)
ax.text(1, -0.1, 'Source: FBREF', transform=ax.transAxes, ha='right', fontsize=12.5)

# Save the chart in high resolution
fig.savefig('ManUtd Radadr.png', dpi=150)

# show the plot
plt.show()
```
